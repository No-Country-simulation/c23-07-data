## Limpieza y pre análisis de los datos 


```python
# Importación de librerias 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Descargamos el archivo parquet y lo pasamos a un dataset
Inflacion = pd.read_parquet(r'C:\Users\LIBARDO\Libardo A\Contry\Mercado financiero\c23-07-data\data\api_data\inflacion.parquet')
Inflacion.head() 
```

```python
# Información basica 
Inflacion.describe()
```

```python
# Información de nulos, vacios y tipo de datos
Inflacion.info()
```

```python
# Transformación de tipo de datos
Inflacion['date'] = pd.to_datetime(Inflacion['date'])
Inflacion['value'] = pd.to_numeric(Inflacion['value'])
```

```python
# Agrupa por mes para que sea se pueda ver mas unificada la información
Inflacion_resampled = Inflacion.resample('M', on='date').mean()

# Graficar los datos agrupados
plt.figure(figsize=(10, 6))
sns.lineplot(data=Inflacion_resampled, x='date', y='value')
plt.ylabel('Inflación')
plt.xlabel('Año')
plt.title('Comportamiento de la inflación a lo largo del tiempo')
plt.show()
```

En el grafico anterior podemos ver un alto indice de inflación que ronda entre los años 1970 a 1980, y en los ultimos años podemos ver un pico considerable.
